```python
sns.barplot(
    x='Machine failure',
    y='Torque [Nm]',
    data=df
)
plt.show()
```

```python
df['Temp_Diff'] = (
    df['Process temperature [K]']
    -
    df['Air temperature [K]']
)
```

```python
sns.barplot(
    x='Machine failure',
    y='Temp_Diff',
    data=df
)
plt.show()
```

```python
sns.barplot(
    x='Machine failure',
    y='Rotational speed [rpm]',
    data=df
)
plt.show()
```

```python
sns.pairplot(
    df[['Air temperature [K]',
        'Process temperature [K]',
        'Rotational speed [rpm]',
        'Torque [Nm]',
        'Machine failure']],
    hue='Machine failure'
)
plt.show()
```
